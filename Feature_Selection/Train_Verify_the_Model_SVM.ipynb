```python
# seleced column: Index(['App Usage Time (min/day)', 'Screen On Time (hours/day)',
#        'Battery Drain (mAh/day)', 'Number of Apps Installed',
#        'Data Usage (MB/day)'],

X_indep=data[['App Usage Time (min/day)','Screen On Time (hours/day)','Battery Drain (mAh/day)', 'Number of Apps Installed','Data Usage (MB/day)']]
```

```python
X_indep.columns
```

```python
Y_dep=data['User Behavior Class']
```

```python
Y_dep
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_indep,Y_dep,test_size=0.30,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
```

```python
param_grid = {'kernel':['linear','rbf','poly','sigmoid'],
             'gamma':['auto','scale'],
             'C':[10,100,1000,2000,3000]} 
grid=GridSearchCV(SVC(probability=True),param_grid,refit=True,verbose=3,cv=5,n_jobs=-1,scoring='f1_weighted')
grid.fit(X_train,Y_train)
```

```python
cv=grid.cv_results_
```

```python
header=pd.DataFrame.from_dict(cv)
```

```python
header
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
Y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,roc_auc_score
```

```python
cm=confusion_matrix(Y_test,Y_pred)
```

```python
cm
```

```python
clasrepo=classification_report(Y_test,Y_pred)
```

```python
load_modle=pickle.load(open('Mobile_usage_SVM_besk_model.sav','rb'))
```

```python
load_modle.feature_names_in_
```

```python
get_inputvalue = []

for i in load_modle.feature_names_in_:
    
    if i=='App Usage Time (min/day)':
        print(f"Enter App Usage Time for {i}:")
        print("Options:0-max (numeric value)")
        model = int(input("Enter the corresponding number: "))
        get_inputvalue.append(model)
    if i=='Screen On Time (hours/day)':
        print(f"Enter Screen On Time for {i}:")
        print("Options:0-max (numeric value)")
        model = float(input("Enter the corresponding number: "))
        get_inputvalue.append(model)
    if i=='Battery Drain (mAh/day)':
        print(f"Enter Battery Drain for {i}:")
        print("Options:0-max (numeric value)")
        model = int(input("Enter the corresponding number: "))
        get_inputvalue.append(model)
    if i=='Number of Apps Installed':
        print(f"Enter 'Number of Apps Installed for {i}:")
        print("Options:0-max (numeric value)")
        model = int(input("Enter the corresponding number: "))
        get_inputvalue.append(model)
    if i=='Data Usage (MB/day)':
        print(f"Enter Data Usage for {i}:")
        print("Options:0-max (numeric value)")
        model = int(input("Enter the corresponding number: "))
        get_inputvalue.append(model)
    
print(get_inputvalue)
```

```python
newpredict=load_modle.predict([get_inputvalue])
```
