```python
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
```

```python
def select_kbest(indped_X,dep_Y,n):
    kbest=SelectKBest(score_func=chi2,k=n)
    selectKfeature=kbest.fit_transform(indped_X,dep_Y)
    mask = kbest.get_support()
    selected_column=indped_X.columns[mask]
    print('seleced column:',selected_column)
    return selectKfeature,selected_column
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
from sklearn.linear_model import LogisticRegression
def logistics_classify(X_train,X_test,Y_train,Y_test):
    classifier=LogisticRegression(random_state=0)
    classifier.fit(X_train,Y_train)
    Y_pred=classifier.predict(X_test)
    acc=accuracy_score(Y_test,Y_pred)
    return acc
```

```python
from sklearn.svm import SVC
def svc_linear_classify(X_train,X_test,Y_train,Y_test):
    classifier=SVC(kernel='linear')
    classifier.fit(X_train,Y_train)
    Y_pred=classifier.predict(X_test)
    acc=accuracy_score(Y_test,Y_pred)
    return acc
```

```python
from sklearn.svm import SVC
def svc_nonlinear_classify(X_train,X_test,Y_train,Y_test):
    classifier=SVC(kernel='rbf',random_state=0)
    classifier.fit(X_train,Y_train)
    Y_pred=classifier.predict(X_test)
    acc=accuracy_score(Y_test,Y_pred)
    return acc
```

```python
from sklearn.naive_bayes import GaussianNB
def gauss_classify(X_train,X_test,Y_train,Y_test):
    classifier=GaussianNB()
    classifier.fit(X_train,Y_train)
    Y_pred=classifier.predict(X_test)
    acc=accuracy_score(Y_test,Y_pred)
    return acc
```

```python
from sklearn.neighbors import KNeighborsClassifier
def KNN_classify(X_train,X_test,Y_train,Y_test):
    classifier=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
    classifier.fit(X_train,Y_train)
    Y_pred=classifier.predict(X_test)
    acc=accuracy_score(Y_test,Y_pred)
    return acc
```

```python
from sklearn.tree import DecisionTreeClassifier
def decision_classify(X_train,X_test,Y_train,Y_test):
    classify=DecisionTreeClassifier(criterion='entropy',random_state=0)
    classify.fit(X_train,Y_train)
    Y_pred=classify.predict(X_test)
    acc=accuracy_score(Y_test,Y_pred)
    return acc
```

```python
from sklearn.ensemble import RandomForestClassifier
def random_classify(X_train,X_test,Y_train,Y_test):
    classify=RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
    classify.fit(X_train,Y_train)
    Y_pred=classify.predict(X_test)
    acc=accuracy_score(Y_test,Y_pred)
    return acc
```

```python
def accurecy_score_in_table(accurecy_score_in_list):
    dataframe=pd.DataFrame(index=['Accuracy'], columns=['Logistic','SVMl','SVMnl','KNN','Navie','Decision','Random'])
    for classifier, score in zip(dataframe.columns,accurecy_score_in_list):
        dataframe.loc['Accuracy', classifier] = score
    return dataframe
```
