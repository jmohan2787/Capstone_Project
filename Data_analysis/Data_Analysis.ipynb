```python
import matplotlib.pyplot as plt
label_mapping = {1: "1-Light Usage",2: "2-Moderate Usage",3: "3-Regular Usage",4: "4-High Usage",5: "5-Extreme Usage"}
plt.pie(data['User Behavior Class'].value_counts(),labels=data['User Behavior Class'].value_counts().index.map(label_mapping),autopct='%1.1f%%')
plt.title('Distribution of User Behavior Classes')
# plt.tight_layout()
# plt.show()
```
